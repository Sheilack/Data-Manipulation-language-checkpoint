# Company Sales Data

You are provided with a company sales data in a file, read the data into a csv file and perform the following activities:

1. Clean up the data, replace all null values in the total sales column with the column average.
2. Create a function which takes in one argument called total_sales then returns 10 % of the total sales value.
3. Create a new column called commission which contains the value generated by the function you created on step 2.
4. Using a pie chart, plot monthly total_sales.
5. Using an histogram, plot sales distribution.
6. Plot a line graph showing monthly sales.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
sales_df = pd.read_csv("sales_data.csv")
sales_df.head()
```

```python
sales_df.info()
```

```python
# replace all null values in the total sales column with the column average.

average_sales = sales_df["total_sales"].mean()
# sales_df["total_sales"].fillna(average_sales, inplace=True)
sales_df.fillna({"total_sales": average_sales}, inplace = True)
sales_df.head(20)
```

```python
# Create a function which takes in one argument called total_sales
# then returns 10 % of the total sales value.

# Create a new column called commission which contains the value generated by the function you created on step 2.

def calculate_commission_earnings(total_sales):
    return total_sales * 0.1
sales_df["commission"] = sales_df["total_sales"].apply(calculate_commission_earnings)
sales_df.head(50)
```

```python
# Using a pie chart, plot monthly total_sales.

monthly_total_sales_counts = sales_df["total_sales"].value_counts()
plt.pie(monthly_total_sales_counts, labels=monthly_total_sales_counts.index, autopct="%1.1f%%")
plt.title("Sales distribution per month")
plt.show()
```
